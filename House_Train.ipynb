```python
import pandas as pd
```

```python
train = pd.read_csv(r'Dataset\train.csv')
train['YearBuilt'] = pd.to_datetime(train['YearBuilt'], format='%Y')
train['YearRemodAdd'] = pd.to_datetime(train['YearRemodAdd'], format='%Y')
train['YrSold'] = pd.to_datetime(train['YrSold'], format='%Y')
train
```

```python
train.info()
```

```python
train.drop(['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'], axis = 1, inplace=True)
```

```python
train
```

```python
from pycaret.regression import *
```

```python
s = setup(data = train, target = 'SalePrice', imputation_type='iterative', numeric_imputation='mean', 
train_size=0.95, categorical_imputation='mode', numeric_iterative_imputer='et', 
categorical_iterative_imputer='et', silent=True)
```

```python
compare_models()
```

```python
lightgbm = create_model('lightgbm')
```

```python
tuned_model = tune_model(lightgbm, optimize='RMSE', choose_better=True)
```

```python
testdata = pd.read_csv('cleantest.csv')
```

```python
pred = predict_model(tuned_model, data=testdata)
```

```python
pred
```

```python
sub = pred[['Id', 'Label']]
sub
```

```python
submission = sub.rename(columns={'Label': 'SalePrice'})
```

```python
submission
```

```python
submission.to_csv('submission.csv', header=True, index=False)
```
