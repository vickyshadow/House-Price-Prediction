```python
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv(r'Dataset\test.csv')
test['YearBuilt'] = pd.to_datetime(test['YearBuilt'], format='%Y')
test['YearRemodAdd'] = pd.to_datetime(test['YearRemodAdd'], format='%Y')
test['YrSold'] = pd.to_datetime(test['YrSold'], format='%Y')
test
```

```python
null_var = test.isna().sum() / test.shape[0] * 100
null_var
```

```python
test.drop(['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'], axis = 1, inplace=True)
```

```python
test.info()
```

```python
fix = null_var[null_var > 0]
fix
```

```python
num_var = test.select_dtypes(include=np.number).columns
num_var
```

```python
for columns in num_var:
    test[columns].fillna(test[columns].mean(), inplace=True)
```

```python
col_var = test.select_dtypes(include='object').keys()
col_var
```

```python
for columns in col_var:
    test[columns].fillna(test[columns].mode().iloc[0], inplace=True)
```

```python
test.isna().sum().sum()
```

```python
test.to_csv('cleantest.csv', index=False, header=True)
```
